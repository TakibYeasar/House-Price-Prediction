```python
categorical_cols = data.select_dtypes(include=['object']).columns

if not categorical_cols.empty:
    num_cols = min(3, len(categorical_cols))
    num_rows = -(-len(categorical_cols) // num_cols)

    plt.figure(figsize=(5 * num_cols, 5 * num_rows))

    for i, col in enumerate(categorical_cols, 1):
        plt.subplot(num_rows, num_cols, i)
        data[col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
        plt.title(f'{col} Distribution')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
else:
    print("No categorical columns found.")
```

```python
numerical_list = data.select_dtypes(include=['int64','float64']).columns.tolist()
numerical_list
```

```python
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns.tolist()

if 'price' in numerical_cols:
    numerical_cols.remove('price')

if numerical_cols:
    num_cols = min(3, len(numerical_cols))
    num_rows = -(-len(numerical_cols) // num_cols)

    plt.figure(figsize=(5 * num_cols, 5 * num_rows))

    for i, col in enumerate(numerical_cols, 1):
        plt.subplot(num_rows, num_cols, i)
        plt.scatter(data[col], data['price'], alpha=0.6, color='royalblue', edgecolors='black')
        plt.title(f'{col} vs Price', fontsize=12, fontweight='bold')
        plt.xlabel(col, fontsize=10)
        plt.ylabel('Price', fontsize=10)
        plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()
else:
    print("No numerical columns found.")
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = data.copy()

categorical_features = df_encoded.select_dtypes(include=['object']).columns

label_encoder = LabelEncoder()

for col in categorical_features:
    unique_values = df_encoded[col].nunique()

    if unique_values == 2:
        # One-Hot Encoding for binary columns
        df_encoded[col] = pd.get_dummies(df_encoded[col], drop_first=True)
    
    elif 3 <= unique_values < 17:
        # Dummy Encoding for categorical variables with moderate cardinality
        df_encoded = pd.get_dummies(df_encoded, columns=[col], drop_first=True, prefix=col)
    
    else:
        # Label Encoding for high-cardinality categorical features
        df_encoded[col] = label_encoder.fit_transform(df_encoded[col])

print(f"Updated dataset shape: {df_encoded.shape}")
```

### Handle Outliers

```python
df_cleaned = df_encoded.copy()
numerical_features = df_cleaned.select_dtypes(include=['number']).columns

def plot_boxplot(df, title):
    plt.figure(figsize=(15, 6))
    df[numerical_features].boxplot(rot=90)
    plt.title(title)
    plt.show()

plot_boxplot(df_cleaned, "Box Plot of Numerical Features Before Outlier Handling")
```

```python
def handle_outliers(df, cap_outliers=False):
    for col in numerical_features:
        Q1, Q3 = df[col].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

        if cap_outliers:
            df[col] = np.clip(df[col], lower, upper)
        else:
            df = df[(df[col] >= lower) & (df[col] <= upper)]

    return df.reset_index(drop=True)

df_cleaned = handle_outliers(df_cleaned, cap_outliers=False)

plot_boxplot(df_cleaned, "Box Plot of Numerical Features After Outlier Handling")

print(f"Before outlier handling: {df_encoded.shape[0]} samples")
print(f"After outlier handling: {df_cleaned.shape[0]} samples")
```

### Split the Data

```python
from sklearn.model_selection import train_test_split  

target = 'price'  

X = df_cleaned.drop(columns=[target])  
y = df_cleaned[target]  

X_train, X_test, y_train, y_test = train_test_split(  
    X, y, test_size=0.2, random_state=52, stratify=y if y.nunique() < 10 else None  
)  
```

### Feature Scaling (Normalization or Standardization)

```python
from sklearn.preprocessing import StandardScaler

X_train_cleaned = X_train.copy()  
X_test_cleaned = X_test.copy()  

scaler = StandardScaler()  
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train_cleaned), columns=X_train_cleaned.columns)  
X_test_scaled = pd.DataFrame(scaler.transform(X_test_cleaned), columns=X_test_cleaned.columns)  

print("\n\033[1mTraining Set (Standardized):\033[0m")  
display(X_train_scaled.describe().round(2))  

print("\n\033[1mTesting Set (Standardized):\033[0m")  
display(X_test_scaled.describe().round(2))
```

### Model Building & Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
}

# Train models
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    print(f"{name} model trained successfully!")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score

# Evaluate models
for name, model in models.items():
    y_pred = model.predict(X_test_scaled)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"\n{name} Model Performance:")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"R² Score: {r2:.2f}")

    # Cross-validation score
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2')
    print(f"Cross-validation R²: {cv_scores.mean():.2f} ± {cv_scores.std():.2f}")
```
